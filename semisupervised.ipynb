```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.svm import SVC
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
import os
import glob


path = os.getcwd()
csv_files = glob.glob(os.path.join(path + '/fiveyear', "*.csv"))
#csv_files = glob.glob(os.path.join(path + '/testfolder', "*.csv"))
#print(path)
#print(csv_files)

compiled_df = []
for f in csv_files:

    if "-Data" in f:
        # read the csv file
        cols = list(pd.read_csv(f, nrows=1))
        #print(cols)

        #df = pd.read_csv(f,  usecols =[i for i in cols if i != "GEO_ID" and i!= "NAME"])
        #df = pd.read_csv(f,  usecols =[i for i in cols if i != "GEO_ID" and i!= "NAME" and "M" not in i and "EA" not in i])
        df = pd.read_csv(f,  usecols =[i for i in cols if i != "GEO_ID" and i!= "NAME" and "M" not in i and "EA" not in i])
        compiled_df.append(df.fillna(0))
        # print the location and filename
        #print('Location:', f)
        #print('File Name:', f.split("\\")[-1])
          
        # print the content
        #print('Content:')
        #display(df)
        #print()

#merged_df =  pd.merge(pd.merge(left,right,on='NAME'), dfs)
frame = pd.concat(compiled_df,  axis=1, join='inner')
display(frame)
```

```python
cleaned_data = frame.tail(-1).replace(to_replace=["*****", "(X)", "-", "**","***","N"],
           value=0).replace(to_replace=["2,500-"],
           value=2500)
#
#df.replace(0,np.nan).dropna(axis=1,how="all")
#
cleaned_data = StandardScaler().fit_transform(cleaned_data)
labels = pd.read_csv("labels.csv", usecols=["IsGentrified?"])
#cleaned_data.target = labels
```

```python
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.svm import SVC
from sklearn.semi_supervised import LabelPropagation
from sklearn.semi_supervised import LabelSpreading
from sklearn import preprocessing
import sklearn
import numpy

svc = SVC(probability=True, class_weight='balanced')

self_training_model = SelfTrainingClassifier(base_estimator=svc)

self_training_model.fit(cleaned_data, labels.values.flatten())

pred = np.array(self_training_model.predict(cleaned_data))
print(pred)

"""
def rbf_kernel_safe(X, Y=None, gamma=None): 

      X, Y = sklearn.metrics.pairwise.check_pairwise_arrays(X, Y) 
      if gamma is None: 
          gamma = 1.0 / X.shape[1] 

      K = sklearn.metrics.pairwise.euclidean_distances(X, Y, squared=True) 
      K *= -gamma 
      K -= K.max()
      np.exp(K, K)    # exponentiate K in-place 
      return K 

model = LabelPropagation(kernel = rbf_kernel_safe, tol = 0.01, gamma = 10)
#model = LabelPropagation()

#model = LabelSpreading(kernel = rbf_kernel_safe, tol = 0.01, gamma = 20)
model.fit(cleaned_data, labels.values.flatten())

pred = np.array(model.predict(cleaned_data))
print(pred)
"""
```
